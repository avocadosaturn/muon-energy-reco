```python
import h5py
import sys
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from scipy.interpolate import splprep, splev
```

```python
f = h5py.File('/sdf/home/s/seohyeon/summer25/test_data/muon/muon1000ev_0-1gev_gampix_5mmpp.h5')
pixels = f['pixel_hits']
coarse = f['coarse_hits']
meta = f['meta']

ev_n = 2

ev_mask_pixels = pixels['event id'] == ev_n
ev_pixels = pixels[ev_mask_pixels]

points = np.array([[hit['pixel x'], hit['pixel y'], hit['hit z']] for hit in ev_pixels])

tck, u = splprep([points[:, 0], points[:, 1], points[:, 2]], s=1000000000)  # s is smoothing factor; s=0 gives interpolation
u_fine = np.linspace(0, 1, 1000)
x_fine, y_fine, z_fine = splev(u_fine, tck)
```

```python
fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(111, projection='3d')


ax.scatter(points[:,0], points[:,1], points[:,2], 'o', s=0.5, alpha =0.1)
ax.plot(x_fine, y_fine, z_fine, 'b-', label='Spline')
```
