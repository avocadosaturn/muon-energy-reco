```python
import h5py
import sys
import numpy as np
import statistics
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import seaborn as sns
import pandas as pd
import os
from scipy.interpolate import make_smoothing_spline, make_splrep, interp1d, BSpline
```

```python
import importlib
import tools
importlib.reload(tools)
from tools import bin_data  


file_ns = range(54)

energies = np.empty(0)
naive = np.empty(0)
detailed = np.empty(0)
charge = np.empty(0)
evs = np.empty(0)


for n in file_ns:
    data = np.load(f'/sdf/data/neutrino/summer25/seohyeon/gampix_estimators/muon1k_0-1gev_estimators_gampix_run{n}.npz')
    

    energies = np.append(energies, data['energies'])
    naive = np.append(naive, data['naive'])
    detailed = np.append(detailed, data['detailed'])
    charge = np.append(charge, data['charge'])
    evs = np.append(evs, data['evs'])
```

```python
splines = np.load('/sdf/data/neutrino/summer25/seohyeon/efxns/efxns_54k_gampix.npz')

naive_spl = BSpline(splines['naive_t'], splines['naive_c'], splines['naive_k'])
detailed_spl = BSpline(splines['detailed_t'], splines['detailed_c'], splines['detailed_k'])



def signed_log(val):
    return np.sign(val) * np.log1p(np.abs(val))  # log1p(val) = log(1 + val)
```

```python
naive_res = naive_spl(naive) - energies
naive_rres = naive_res / energies

detailed_res = detailed_spl(detailed) - energies
detailed_rres = detailed_res / energies

global_evs = np.empty(0)
prev = 0
mult = 0
for ev in evs:
    current = ev
    if current - prev < 0:
        mult += 1000
        global_evs = np.append(global_evs, mult + current)
        prev = ev
    else:
        global_evs = np.append(global_evs, mult + current)
        prev = ev


outliers = [1320, 19956, 3780, 49749, 28732, 1463, 1833, 2012, 2945, 3358, 5288, 10772, 11387, 12611, 12680, 14093, 15212, 16018, 19101, 19345, 19427, 19941, 22016, 22658, 23485, 24292, 24733, 25839, 26498, 27067, 28535, 28801, 29148, 29493, 30028, 30258, 30575, 32519, 32532, 33498, 33509, 35518, 35753, 36699, 37155, 37304, 37796, 37843, 38281, 39747, 40708, 41005, 41940, 44496, 46371, 48920, 49876, 50547, 50756, 52289, 52627, 53064, 53066, 53167]
outliersg = []
for outlier in outliers:
    i = np.where(global_evs == outlier)
    if len(i[0]) == 0:
        continue
    else:
        outliersg.append(i[0][0])

energies = np.delete(energies, outliersg)
naive_rres = np.delete(naive_rres, outliersg)
detailed_rres = np.delete(detailed_rres, outliersg)
```

```python
naive_rres_x, naive_rres_y, naive_rres_error, naive_rres_rrms = bin_data(energies, naive_rres, 40)

# plot of all data and binned data with no log scale
plt.figure(figsize=(8, 5))
# plt.ylim(-0.15, 0.1)
plt.plot(energies, naive_rres, 'o', ms=4, color='lightgray')
plt.errorbar(naive_rres_x, naive_rres_y, naive_rres_error, fmt='o', color='black', ms=5)
plt.xlabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.ylabel(r'$\frac{T_\text{n} - T_\mu}{T_\mu}$', fontsize=12)
plt.title(f'naive length relative residual vs. muon KE', fontsize=16)
# plt.savefig(f'Plots/accuracy_no_DIF/naive_rres_{total_evs}evs.png')


# plot of all data and binned data on log scale 
plt.figure(figsize=(8, 5))
plt.ylim(-0.2, 0.2)
plt.plot(energies, signed_log(naive_rres), 'o', ms=4, color='lightgray')
plt.errorbar(naive_rres_x, signed_log(naive_rres_y), signed_log(naive_rres_error), fmt='o', color='black', ms=5)
plt.xlabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.ylabel(r'signed log($\frac{T_\text{n} - T_\mu}{T_\mu}$)', fontsize=12)
plt.title(f'naive length relative residual vs. muon KE', fontsize=16)
# plt.savefig(f'Plots/accuracy_no_DIF/naive_rres_log_{total_evs}evs.png')
```

```python
detailed_rres_x, detailed_rres_y, detailed_rres_error, detailed_rres_rrms = bin_data(energies, detailed_rres, 40)

# plot of all data and binned data with no log scale
plt.figure(figsize=(8, 5))
plt.plot(energies, detailed_rres, 'o', ms=4, color='lightgray')
# plt.plot(energies[10770], detailed_rres[10770], 'o', color='blue')
plt.errorbar(detailed_rres_x, detailed_rres_y, detailed_rres_error, fmt='o', color='black', ms=5)
plt.xlabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.ylabel(r'$\frac{T_\text{d} - T_\mu}{T_\mu}$', fontsize=12)
plt.title(f'detailed length relative residual vs. muon KE', fontsize=16)
# plt.savefig(f'Plots/accuracy_no_DIF/detailed_rres_{total_evs}evs.png')


# plot of all data and binned data on log scale 
plt.figure(figsize=(8, 5))
plt.ylim(-0.2, 0.2)
plt.plot(energies, signed_log(detailed_rres), 'o', ms=4, color='lightgray')
plt.errorbar(detailed_rres_x, signed_log(detailed_rres_y), signed_log(detailed_rres_error), fmt='o', color='black', ms=5)
plt.xlabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.ylabel(r'signed log($\frac{T_\text{d} - T_\mu}{T_\mu}$)', fontsize=12)
plt.title(f'detailed length relative residual vs. muon KE', fontsize=16)
# plt.savefig(f'Plots/accuracy_no_DIF/detailed_rres_log_{total_evs}evs.png')
```
