```python
import h5py
import sys
import numpy as np
import matplotlib.pyplot as plt
from tools import bin20
from sklearn.decomposition import PCA
```

```python
ranges=[] #cm
energies=[]

f = h5py.File('/sdf/home/s/seohyeon/summer25/data/muon/mubar_1000_0-1GeV_gampixsim_5mmpp.h5')
pixels = f['pixel_hits']
tiles = f['coarse_hits']
meta = f['meta']

evs = meta['event id']

for ev_n in evs:
    # print('---------------------')
    # print(f'event {ev_n}')
    ev_mask_tiles = tiles['event id'] == ev_n
    ev_tiles = tiles[ev_mask_tiles]
    
    #getting the coordinates of each hit
    coords = []
    for hit in ev_tiles:
        hit_coord = np.array([hit['tile x'], hit['tile y'], hit['hit z']])
        coords.append(hit_coord)
    # print(f'number of pts: {len(coords)}')

    
    # pca analysis
    min_hits = 100
    if len(coords) < min_hits:
        continue
    else:
        pca = PCA(n_components = 3)
        pca_on_hits = pca.fit(coords)
        principal=pca.components_[0] #this is a unit vector

    KE = meta['primary energy'][ev_n]
    energies.append(KE)


    #getting projection of each hit onto the principal axis
    parallel_proj={} #key is magnitude of projection, value is coordinate associated with it
    antiparallel_proj={}
    for coord in coords:
        proj = np.dot(principal, coord) * principal
        mag_proj = np.linalg.norm(proj)
    
        cos = np.dot(principal, proj) / mag_proj #cos theta between the projection and principal axis. Will be 1 if parallel, -1 if antiparallel
        if cos > 0:
            parallel_proj[mag_proj] = coord
        elif cos < 0:
            antiparallel_proj[mag_proj] = coord


    # getting track length
    if len(parallel_proj)==0:
        start=antiparallel_proj[min(antiparallel_proj)]
        stop=antiparallel_proj[max(antiparallel_proj)]
        track_length=np.linalg.norm(start-stop)
    elif len(antiparallel_proj)==0:
        start=parallel_proj[min(parallel_proj)]
        stop=parallel_proj[max(parallel_proj)]
        track_length=np.linalg.norm(start-stop)
    else:
        start=antiparallel_proj[max(antiparallel_proj)]
        stop=parallel_proj[max(parallel_proj)]
        track_length=max(parallel_proj)+max(antiparallel_proj)
    ranges.append(track_length)

    # if KE > 700 and track_length < 6750:
    #     print('-------')
    #     print(ev_n)                                                                                                                                                                                                                                                                                                                                                  

# print(len(energies))
# print(len(ranges))

plt.figure()
plt.plot(energies, ranges, 'o')
plt.xlabel('muon KE (MeV)')
plt.ylabel('range (cm)')
plt.title(f'range vs. muon KE with coarse tiles, {min_hits} hit threshold')
plt.savefig(f'Plots/coarse_range_{max(evs)+1}evs_{min_hits}threshold.png')
```
