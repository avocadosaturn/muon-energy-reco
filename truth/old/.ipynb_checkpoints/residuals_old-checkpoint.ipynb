```python
import h5py
import sys
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import os
from scipy.interpolate import make_smoothing_spline, make_splrep, interp1d
```

```python
import importlib
import tools
importlib.reload(tools)

from tools import bin_data  

data = np.load('muon_0-1gev_10k.npz')

energies = data['energies']
ranges = data['ranges']
path_lengths = data['path_lengths']
deposited_Es = data['deposited_Es']
evs = data['evs']

total_evs = '10k'
```

curve paramaterizations:

```python
#range (piecewise fit) -------------------------------
ranges = np.array(ranges)
range_bin_midpts, range_avgs, range_errors, range_rel_rms = bin_data(ranges, energies, 40)
# to prevent the binning curve
range_bin_midpts = range_bin_midpts[:-4]
range_avgs = range_avgs[:-4]


range_bin_midpts = np.insert(range_bin_midpts, 0, np.min(ranges))
range_avgs = np.insert(range_avgs, 0, energies[np.argmin(ranges)])


range_spl = make_splrep(range_bin_midpts, range_avgs, s=30)

plt.figure()
# plt.plot(ranges_new, range_energies, 'o', ms=4, color='lightgray')
plt.plot(range_bin_midpts, range_avgs, 'o', ms=5, color='black')
# plt.plot(range_bin_midpts, range_yfit_display, color='black', linewidth=1)
# plt.plot(range_bin_midpts, range_yfit(range_bin_midpts), linewidth=1, color='black')
plt.plot(range_bin_midpts, range_spl(range_bin_midpts), linewidth=1, color='black')
plt.xlabel(r'$r_n$ (mm)', fontsize=12)
plt.ylabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.title(f'true muon KE vs. naive length, {total_evs} events', fontsize=16)
plt.savefig(f'Plots/residuals/range_fit_{total_evs}evs.png')
```

```python
# path length -----------------------------------
path_lengths = np.array(path_lengths)
pl_bin_midpts, pl_avgs, pl_errors, pl_rel_rms = bin_data(path_lengths, energies, 40)
pl_bin_midpts = pl_bin_midpts[:-4]
pl_avgs = pl_avgs[:-4]

pl_bin_midpts = np.insert(pl_bin_midpts, 0, np.min(path_lengths))
pl_avgs = np.insert(pl_avgs, 0, energies[np.argmin(path_lengths)])

pl_spl = make_splrep(pl_bin_midpts, pl_avgs, s=30)

plt.figure()
# plt.plot(path_lengths, energies, 'o', color='lightgray')
plt.plot(pl_bin_midpts, pl_avgs, 'o', color='black', ms=5)
# plt.plot(pl_bin_midpts, pl_yfit_display, color='black', linewidth=1)
plt.plot(pl_bin_midpts, pl_spl(pl_bin_midpts), linewidth=1, color='black')
plt.xlabel(r'$r_d$ (mm)', fontsize=12)
plt.ylabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.title(f'true muon KE vs. detailed track length, {total_evs} events', fontsize=16)
plt.savefig(f'Plots/residuals/pl_fit_{total_evs}evs.png')
```

```python
# edep ---------------------------------------------------------
deposited_Es = np.array(deposited_Es)
edep_bin_midpts, edep_avgs, edep_rms, edep_relative_rms = bin_data(deposited_Es, energies, 40)
edep_bin_midpts = edep_bin_midpts[:-4]
edep_avgs = edep_avgs[:-4]

edep_bin_midpts = np.insert(edep_bin_midpts, 0, np.min(deposited_Es))
edep_avgs = np.insert(edep_avgs, 0, energies[np.argmin(deposited_Es)])


edep_spl = make_splrep(edep_bin_midpts, edep_avgs, s=30)


plt.figure()
# plt.plot(deposited_Es, energies, 'o', ms=4, color='lightgray')
plt.plot(edep_bin_midpts, edep_avgs, 'o', color='black', ms=5)
# plt.plot(edep_bin_midpts, edep_yfit_display, color='black', linewidth=1)
plt.plot(edep_bin_midpts, edep_spl(edep_bin_midpts), linewidth=1, color='black')
plt.xlabel(r'$E_\text{dep}$ (MeV)', fontsize=12)
plt.ylabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.title(f'true muon KE vs. energy deposited, {total_evs} events', fontsize=16)
plt.savefig(f'Plots/residuals/edep_fit_{total_evs}evs.png')
```

looking for complmentary info:

```python
range_res = range_spl(ranges) - energies
range_relative_res = range_res / energies

pl_res = pl_spl(path_lengths) - energies
pl_relative_res = pl_res / energies

edep_res = edep_spl(deposited_Es) - energies
edep_relative_res = edep_res / energies
```

```python
plt.figure(figsize=(8, 6))
plt.hist2d(edep_relative_res, range_relative_res, bins=60, range=[[-0.3, 0.3], [-0.3, 0.3]])
plt.xlabel(r'$\frac{T_\text{edep} - T_\mu}{T_\mu}$', fontsize=12)
plt.ylabel(r'$\frac{T_\text{n} - T_\mu}{T_\mu}$', fontsize=12)
plt.title(f'naive length relative residual vs. edep relative residual, {total_evs} evs', fontsize=16)
plt.savefig(f'Plots/residuals/rangevsedep_rres_{total_evs}evs.png')

plt.figure(figsize=(8, 6))
plt.hist2d(edep_relative_res, pl_relative_res, bins=60, range=[[-0.3, 0.3], [-0.3, 0.3]])
plt.xlabel(r'$\frac{T_\text{edep} - T_\mu}{T_\mu}$', fontsize=12)
plt.ylabel(r'$\frac{T_\text{d} - T_\mu}{T_\mu}$', fontsize=12)
plt.title(f'detailed length relative residual vs. edep relative residual, {total_evs} evs', fontsize=16)
plt.savefig(f'Plots/residuals/plvsedep_rres_{total_evs}evs.png')

plt.figure(figsize=(8, 6))
plt.hist2d(range_relative_res, pl_relative_res, bins=60, range=[[-0.3, 0.3], [-0.3, 0.3]])
plt.xlabel(r'$\frac{T_\text{n} - T_\mu}{T_\mu}$', fontsize=12)
plt.ylabel(r'$\frac{T_\text{d} - T_\mu}{T_\mu}$', fontsize=12)
plt.title(f'detailed length relative residual vs. range relative residual, {total_evs} evs', fontsize=16)
plt.savefig(f'Plots/residuals/plvsrange_rres_{total_evs}evs.png')
```

plots of relative residuals:

```python
def signed_log(val):
    return np.sign(val) * np.log1p(np.abs(val))  # log1p(val) = log(1 + val)

# naive ---------------------
range_rel_res = np.array(range_relative_res)
range_rres_bin_midpts, range_rres_avgs, range_rres_rms, range_rres_rel_rms = bin_data(energies, range_rel_res, 40)

# plot of all data and binned data with no log scale
plt.figure(figsize=(8, 5))
# plt.ylim(-0.15, 0.1)
plt.plot(energies, range_rel_res, 'o', ms=4, color='lightgray')
plt.errorbar(range_rres_bin_midpts, range_rres_avgs, range_rres_rms, fmt='o', color='black', ms=5)
plt.xlabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.ylabel(r'$\frac{T_\text{n} - T_\mu}{T_\mu}$', fontsize=12)
plt.title(f'naive length relative residual vs. muon KE, {total_evs} evs', fontsize=16)
plt.savefig(f'Plots/residuals/range_rres_{total_evs}evs.png')


# plot of all data and binned data on log scale 
plt.figure(figsize=(8, 5))
plt.ylim(-0.1, 0.1)
plt.plot(energies, signed_log(range_rel_res), 'o', ms=4, color='lightgray')
plt.errorbar(range_rres_bin_midpts, signed_log(range_rres_avgs), signed_log(range_rres_rms), fmt='o', color='black', ms=5)
plt.xlabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.ylabel(r'signed log($\frac{T_\text{n} - T_\mu}{T_\mu}$)', fontsize=12)
plt.title(f'naive length relative residual vs. muon KE, {total_evs} evs', fontsize=16)
plt.savefig(f'Plots/residuals/range_rres_log_{total_evs}evs.png')

```

```python
# detailed ----------------
pl_rel_res = np.array(pl_relative_res)
pl_rres_bin_midpts, pl_rres_avgs, pl_rres_rms, pl_rres_rel_rms = bin_data(energies, pl_rel_res, 40)

plt.figure(figsize=(8, 5))
plt.plot(energies, pl_rel_res, 'o', ms=4, color='lightgray')
plt.errorbar(pl_rres_bin_midpts, pl_rres_avgs, pl_rres_rms, fmt='o', ms=5, color='black')
plt.xlabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.ylabel(r'$\frac{T_\text{d} - T_\mu}{T_\mu}$', fontsize=12)
plt.title(f'detailed length relative residual vs. muon KE, {total_evs} evs', fontsize=16)
plt.savefig(f'Plots/residuals/pl_rres_{total_evs}evs.png') 

plt.figure(figsize=(8, 5))
plt.ylim(-0.1, 0.1)
plt.plot(energies, signed_log(pl_rel_res), 'o', ms=4, color='lightgray')
plt.errorbar(pl_rres_bin_midpts, signed_log(pl_rres_avgs), signed_log(pl_rres_rms), fmt='o', color='black', ms=5)
plt.xlabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.ylabel(r'signed log($\frac{T_\text{d} - T_\mu}{T_\mu}$)', fontsize=12)
plt.title(f'detailed length relative residual vs. muon KE {total_evs} evs', fontsize=16)
plt.savefig(f'Plots/residuals/pl_rres_log_{total_evs}evs.png')

```

```python
# edep --------------------
edep_rres = np.array(edep_relative_res)
edep_rres_bin_midpts, edep_rres_avgs, edep_rres_rms, edep_rres_rel_rms = bin_data(energies, edep_rres, 40)


plt.figure(figsize=(8, 5))
plt.plot(energies, edep_rres, 'o', ms=4, color='lightgray')
plt.errorbar(edep_rres_bin_midpts, edep_rres_avgs, edep_rres_rms, fmt='o', ms=5, color='black')
plt.xlabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.ylabel(r'$\frac{T_\text{edep} - T_\mu}{T_\mu}$', fontsize=12)
plt.title('edep relative residual vs. muon KE', fontsize=16)
plt.savefig(f'Plots/residuals/edep_rres_{total_evs}evs.png')


plt.figure(figsize=(8, 5))
plt.ylim(-0.1, 0.1)
plt.plot(energies, signed_log(edep_rres), 'o', ms=4, color='lightgray')
plt.errorbar(edep_rres_bin_midpts, signed_log(edep_rres_avgs), signed_log(edep_rres_rms), fmt='o', ms=5, color='black')
plt.xlabel(r'$T_\mu$ (MeV)', fontsize=12)
plt.ylabel(r'signed log($\frac{T_\text{edep} - T_\mu}{T_\mu}$)', fontsize=12)
plt.title(f'edep relative residual vs. muon KE, {total_evs} evs', fontsize=16)
plt.savefig(f'Plots/residuals/edep_rres_log_{total_evs}evs.png')
```
